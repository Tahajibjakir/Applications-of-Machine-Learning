```python
import pandas as pd
import numpy as np
import matplotlib.pyplot  as plt
```

```python
df=pd.read_csv('carprices.csv')
```

```python
df
```

```python
dummy_var=pd.get_dummies(df['Car Model'],drop_first=True)
```

```python
dummy_var
```

```python
new_df=df.drop('Car Model', axis=1)
```

```python
new_df
```

```python
df=pd.concat([new_df,dummy_var],axis=1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Sell Price',axis=1)
```

```python
x
```

```python
y=df['Sell Price']
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3, random_state=42)
```

```python
xtrain.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lre=LinearRegression()
```

```python
lre.fit(xtrain,ytrain)
```

```python
lre.predict(xtest)
```

```python
xtest.shape
```

```python
lre.score(xtest,ytest)
```

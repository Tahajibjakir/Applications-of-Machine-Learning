```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder
```

```python
label=LabelEncoder()
```

```python
# Loop


for column in df.columns:
   if df[column].dtype==np.number:
       continue
   else:
       df[column]=label.fit_transform(df[column])
```

# Ordinal Encoder

```python
df=pd.read_csv('Area Price.csv')
df
```

```python

from sklearn.preprocessing import OrdinalEncoder
```

```python
df.Area.unique()
```

```python
city_list=['Dhaka', 'Ctg', 'Rangpur']
```

```python
Ordinal=OrdinalEncoder(categories=[city_list])
```

```python
encoded_value= Ordinal.fit_transform(df[['Area']]) #It works with Number of sample and number of features at a time thats why two dimention.
```

```python
encoded_value
```

```python
new_area= pd.DataFrame(encoded_value, columns=['Area'])
```
